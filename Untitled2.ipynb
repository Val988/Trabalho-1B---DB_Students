```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

def validar_extensao(nome_arquivo):
    return nome_arquivo.endswith('.csv') or nome_arquivo.endswith('.json')

def escolher_arquivo():
    while True:
        print("\nEscolha uma opção:")
        print("1 - Fazer upload de um novo arquivo (.csv ou .json)")
        print("2 - Selecionar um arquivo já enviado")

        escolha = input("Digite 1 ou 2: ").strip()

        if escolha == '1':
            print("\nEnvie o arquivo (.csv ou .json)...")
            uploaded = files.upload()

            if not uploaded:
                print("Nenhum arquivo enviado.")
                continue

            nome_arquivo = list(uploaded.keys())[0]

            if not validar_extensao(nome_arquivo):
                print(
                    f"❌ Arquivo '{nome_arquivo}' tem formato inválido. Use .csv ou .json."
                )
                os.remove(nome_arquivo)
                continue

            print(f"✅ Arquivo '{nome_arquivo}' enviado com sucesso.")
            return nome_arquivo

        elif escolha == '2':
            arquivos = [f for f in os.listdir() if validar_extensao(f)]

            if not arquivos:
                print("⚠️ Nenhum arquivo .csv ou .json encontrado. Faça o upload primeiro.")
                continue

            print("\nArquivos disponíveis:")
            for i, nome in enumerate(arquivos):
                print(f"{i + 1}. {nome}")

            try:
                opcao = int(input("Digite o número do arquivo desejado: ")) - 1
                if 0 <= opcao < len(arquivos):
                    return arquivos[opcao]
                print("❌ Opção inválida.")
            except ValueError:
                print("❌ Entrada inválida. Digite apenas números.")

        else:
            print("❌ Escolha inválida. Tente novamente.")

def carregar_dados(caminho_arquivo):
    if caminho_arquivo.endswith('.csv'):
        return pd.read_csv(caminho_arquivo)
    if caminho_arquivo.endswith('.json'):
        return pd.read_json(caminho_arquivo)
    raise ValueError("Formato de arquivo não suportado. Use CSV ou JSON.")

def limpar_dados(df):
    col_educacao = next(
        (
            col for col in df.columns
            if 'parent' in col.lower() and 'education' in col.lower()
        ),
        None
    )
    if col_educacao:
        df = df[df[col_educacao].notna()]

    if 'Attendance' in df.columns:
        mediana_attendance = df['Attendance'].median()
        df['Attendance'].fillna(mediana_attendance, inplace=True)
        print("\nSomatório de Attendance após preenchimento:", df['Attendance'].sum())

    return df

def resumo_dados(df):
    print("\nResumo dos Dados")
    print("-" * 30)
    print("Quantidade total de registros:", len(df))

    colunas_disponiveis = df.columns.tolist()
    print("Colunas disponíveis no dataset:", colunas_disponiveis)

    genero_coluna = next(
        (col for col in colunas_disponiveis if 'gender' in col.lower()), None
    )
    if genero_coluna:
        print("Quantidade de homens:", (df[genero_coluna].str.lower() == 'male').sum())
        print("Quantidade de mulheres:", (df[genero_coluna].str.lower() == 'female').sum())
    else:
        print("Coluna de gênero não encontrada.")

    educacao_coluna = next(
        (
            col for col in colunas_disponiveis
            if 'parent' in col.lower() and 'education' in col.lower()
        ),
        None
    )
    if educacao_coluna:
        print(
            "Registros sem informação sobre a educação dos pais:",
            df[educacao_coluna].isna().sum()
        )
    else:
        print("Coluna de educação dos pais não encontrada.")

def estatisticas_coluna(df):
    colunas_numericas = df.select_dtypes(include='number').columns.tolist()

    if not colunas_numericas:
        print("Não há colunas numéricas disponíveis no dataset.")
        return

    while True:
        print("\nColunas numéricas disponíveis:")
        for i, col in enumerate(colunas_numericas):
            print(f"{i + 1}. {col}")

        try:
            escolha = int(input("\nDigite o número da coluna que deseja analisar: ")) - 1
            if 0 <= escolha < len(colunas_numericas):
                coluna = colunas_numericas[escolha]
                serie = df[coluna].dropna()

                media = serie.mean()
                mediana = serie.median()
                moda = serie.mode()
                desvio_padrao = serie.std()

                print(f"\nEstatísticas para a coluna '{coluna}':")
                print(f"Média: {media}")
                print(f"Mediana: {mediana}")
                print(f"Moda: {moda.values if not moda.empty else 'Sem moda'}")
                print(f"Desvio padrão: {desvio_padrao}")
            else:
                print("Número inválido.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

        continuar = input("\nDeseja analisar outra coluna? (s/n): ").strip().lower()
        if continuar != 's':
            print("Encerrando análise estatística.")
            break

def grafico_dispersao(df):
    x = df['Sleep_Hours_per_Night']
    y = df['Final_Score']

    plt.figure(figsize=(10, 6))
    plt.scatter(
        x, y, color='#2196F3', edgecolor='k', alpha=0.7, s=80, label='Alunos'
    )

    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x, p(x), color='red', linestyle='--', linewidth=2, label='Tendência')

    plt.title(
        'Horas de Sono por Noite x Nota Final', fontsize=14, fontweight='bold'
    )
    plt.xlabel('Horas de Sono por Noite', fontsize=12)
    plt.ylabel('Nota Final', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

def grafico_barras_idade_midterm(df):
    media_por_idade = df.groupby('Age')['Midterm_Score'].mean().sort_index()

    plt.figure(figsize=(10, 6))
    bars = plt.bar(
        media_por_idade.index.astype(str),
        media_por_idade.values,
        color='#4CAF50',
        edgecolor='black'
    )

    plt.title(
        'Idade x Média das Notas Intermediárias', fontsize=14, fontweight='bold'
    )
    plt.xlabel('Idade', fontsize=12)
    plt.ylabel('Média das Notas Intermediárias', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2.0,
            height + 0.5,
            f'{height:.1f}',
            ha='center',
            va='bottom',
            fontsize=10
        )

    plt.show()

def classificar_faixa_etaria(idade):
    if idade <= 17:
        return 'Até 17'
    if 18 <= idade <= 21:
        return '18 a 21'
    if 22 <= idade <= 24:
        return '22 a 24'
    return '25 ou mais'

def grafico_pizza_faixas_etarias(df):
    df['Faixa_Etaria'] = df['Age'].apply(classificar_faixa_etaria)
    distribuicao = df['Faixa_Etaria'].value_counts()

    plt.figure(figsize=(6, 6))
    distribuicao.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title('Distribuição por Faixa Etária')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()

def main():
    caminho_arquivo = escolher_arquivo()

    if caminho_arquivo:
        try:
            df = carregar_dados(caminho_arquivo)
            df = limpar_dados(df)
            resumo_dados(df)
            estatisticas_coluna(df)
            grafico_dispersao(df)
            grafico_barras_idade_midterm(df)
            grafico_pizza_faixas_etarias(df)
        except Exception as e:
            print("Erro ao carregar os dados:", e)

if __name__ == '__main__':
    main()
```
