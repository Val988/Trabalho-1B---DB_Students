```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files


def validar_extensao(nome_arquivo):
    return nome_arquivo.endswith('.csv') or nome_arquivo.endswith('.json')


def escolher_arquivo():
    while True:
        print("\nEscolha uma opção:")
        print("1 - Fazer upload de um novo arquivo (.csv ou .json)")
        print("2 - Selecionar um arquivo já enviado")

        escolha = input("Digite 1 ou 2: ").strip()

        if escolha == '1':
            print("\nEnvie o arquivo (.csv ou .json)...")
            uploaded = files.upload()

            if not uploaded:
                print("Nenhum arquivo enviado.")
                continue

            nome_arquivo = list(uploaded.keys())[0]

            if not validar_extensao(nome_arquivo):
                print(f"❌ Arquivo '{nome_arquivo}' inválido. Use .csv ou .json.")
                os.remove(nome_arquivo)
                continue

            print(f"✅ Arquivo '{nome_arquivo}' enviado com sucesso.")
            return nome_arquivo

        elif escolha == '2':
            arquivos = [f for f in os.listdir() if validar_extensao(f)]

            if not arquivos:
                print("⚠️ Nenhum arquivo .csv ou .json encontrado.")
                continue

            print("\nArquivos disponíveis:")
            for i, nome in enumerate(arquivos):
                print(f"{i + 1}. {nome}")

            try:
                opcao = int(input("Digite o número do arquivo desejado: ")) - 1
                if 0 <= opcao < len(arquivos):
                    return arquivos[opcao]
                print("❌ Opção inválida.")
            except ValueError:
                print("❌ Entrada inválida. Use apenas números.")

        else:
            print("❌ Escolha inválida. Tente novamente.")


def carregar_dados(caminho_arquivo):
    if caminho_arquivo.endswith('.csv'):
        return pd.read_csv(caminho_arquivo)
    if caminho_arquivo.endswith('.json'):
        return pd.read_json(caminho_arquivo)
    raise ValueError("Formato não suportado. Use CSV ou JSON.")


def resumo_dados(df):
    print("\nResumo dos Dados")
    print("-" * 30)
    print("Total de registros:", len(df))

    colunas = df.columns.tolist()
    print("Colunas disponíveis:", colunas)

    genero_col = next((col for col in colunas if 'gender' in col.lower()), None)
    if genero_col:
        masc = (df[genero_col].str.lower() == 'male').sum()
        fem = (df[genero_col].str.lower() == 'female').sum()
        print("Homens:", masc)
        print("Mulheres:", fem)
    else:
        print("Coluna de gênero não encontrada.")

    educa_col = next((col for col in colunas
                      if 'parent' in col.lower() and 'education' in col.lower()),
                     None)
    if educa_col:
        faltando = df[educa_col].isna().sum()
        print("Sem info. sobre educação dos pais:", faltando)
    else:
        print("Coluna de educação dos pais não encontrada.")


def estatisticas_coluna(df):
    col_num = df.select_dtypes(include='number').columns.tolist()

    if not col_num:
        print("Sem colunas numéricas disponíveis.")
        return

    while True:
        print("\nColunas numéricas disponíveis:")
        for i, col in enumerate(col_num):
            print(f"{i + 1}. {col}")

        try:
            esc = int(input("\nEscolha a coluna: ")) - 1
            if 0 <= esc < len(col_num):
                col = col_num[esc]
                serie = df[col].dropna()

                print(f"\nEstatísticas para '{col}':")
                print(f"Média: {serie.mean()}")
                print(f"Mediana: {serie.median()}")
                moda = serie.mode()
                print(f"Moda: {moda.values if not moda.empty else 'Sem moda'}")
                print(f"Desvio padrão: {serie.std()}")
            else:
                print("Número inválido.")
        except ValueError:
            print("Entrada inválida. Use um número.")

        continuar = input("\nDeseja analisar outra coluna? (s/n): ")
        if continuar.strip().lower() != 's':
            break


def grafico_dispersao(df):
    x = df['Sleep_Hours_per_Night']
    y = df['Final_Score']

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='#2196F3', edgecolor='k', alpha=0.7,
                s=80, label='Alunos')

    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x, p(x), color='red', linestyle='--', linewidth=2,
             label='Tendência')

    plt.title('Sono x Nota Final', fontsize=14, fontweight='bold')
    plt.xlabel('Horas de Sono', fontsize=12)
    plt.ylabel('Nota Final', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()


def grafico_barras_idade_midterm(df):
    media_idade = df.groupby('Age')['Midterm_Score'].mean().sort_index()

    plt.figure(figsize=(10, 6))
    bars = plt.bar(media_idade.index.astype(str), media_idade.values,
                   color='#4CAF50', edgecolor='black')

    plt.title('Idade x Nota Intermediária', fontsize=14, fontweight='bold')
    plt.xlabel('Idade', fontsize=12)
    plt.ylabel('Média da Nota Intermediária', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    for bar in bars:
        altura = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2.0, altura + 0.5,
                 f'{altura:.1f}', ha='center', va='bottom', fontsize=10)

    plt.show()


def classificar_faixa_etaria(idade):
    if idade <= 17:
        return 'Até 17'
    if 18 <= idade <= 21:
        return '18 a 21'
    if 22 <= idade <= 24:
        return '22 a 24'
    return '25 ou mais'


def grafico_pizza_faixas_etarias(df):
    df['Faixa_Etaria'] = df['Age'].apply(classificar_faixa_etaria)
    dist = df['Faixa_Etaria'].value_counts()

    plt.figure(figsize=(6, 6))
    dist.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title('Distribuição por Faixa Etária')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()


def main():
    caminho = escolher_arquivo()

    if not caminho:
        return

    try:
        df = carregar_dados(caminho)
    except Exception as e:
        print("Erro ao carregar os dados:", e)
        return

    while True:
        print("\nEscolha uma ação:")
        print("1 - Resumo dos dados")
        print("2 - Estatísticas de colunas numéricas")
        print("3 - Gráfico de dispersão (Sono x Nota Final)")
        print("4 - Gráfico de barras (Idade x Nota Intermediária)")
        print("5 - Gráfico de pizza (Faixa Etária)")
        print("6 - Sair")

        esc = input("Digite o número da opção: ").strip()

        if esc == '1':
            resumo_dados(df)
        elif esc == '2':
            estatisticas_coluna(df)
        elif esc == '3':
            grafico_dispersao(df)
        elif esc == '4':
            grafico_barras_idade_midterm(df)
        elif esc == '5':
            grafico_pizza_faixas_etarias(df)
        elif esc == '6':
            print("Encerrando o programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")


if __name__ == '__main__':
    main()
```
